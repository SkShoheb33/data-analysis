```python
data = pd.read_csv('C:\\Users\\exam2\\Downloads\\movies_1_.csv')

# data
```

```python
#1. 	 (i). Find out the no of movies released in every month of the year 1995.
data.drop(data[data['Release Date']=='TBD'].index,axis=0,inplace=True)

data["Release Date"]=data['Release Date'].apply(lambda x:pd.to_datetime(x))

# data
```

```python
data = data[data['Release Date'].apply(lambda x:x.year)==1995]
data.groupby(pd.Grouper(key='Release Date',axis=0,freq='m'))['Title'].count()
```

```python
#(ii). Find out the no. of movies released in every year from 1990 to 1998.
data = data[data['Release Date'].apply(lambda x:x.year).between(1990,1998)]
data.groupby(pd.Grouper(key='Release Date',axis=0,freq='y'))['Title'].count()
```

```python
# 2.	(i). Find no. of movies released under each genre given in the database.
key = 'Major Genre'
data.groupby([key])['Title'].count()
```

```python
#(ii). Find the movies under each genre with 1MDB rating >7 and rotten tomatoes rating > 60.

data[(data['IMDB Rating']>7) & (data['Rotten Tomatoes Rating']>60)].groupby(['Major Genre'])['Title'].count()
```

```python
data = data.fillna(0)
```
